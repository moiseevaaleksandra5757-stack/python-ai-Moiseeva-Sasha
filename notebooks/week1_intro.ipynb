## ⚡ Первый пример: логические переменные

**Логическая переменная (bool)** может иметь только два значения: `True` или `False`.

Попробуйте изменить значение `E` на `False` и посмотрите, что произойдёт!

**Обратите внимание** на использование оператора `and` во втором примере — он проверяет ОБА условия одновременно.

````python
# Простой пример
E = True  # Энергия
if E:
    print("⚡ Чайник включён!")
else:
    print("❌ Чайник выключен")

print()  # Пустая строка для разделения

# Сложный пример
E = True   # Есть энергия
W = True   # Есть вода

if E and W:
    print("✅ Чайник включён и вода нагревается!")
elif E and not W:
    print("⚠️ Чайник включён, но воды нет!")
elif not E and W:
    print("⚠️ Вода есть, но нет электричества!")
else:
    print("❌ Нет ни воды, ни электричества!")
````

## 👥 Работа со списками: сказка про репку

Здесь мы видим:
- Создание списка с эмодзи
- Срез списка `heroes[1:]` — все элементы, кроме первого
- Цикл `for` для перебора элементов
- Конкатенацию строк через `+=`

**Попробуйте:** добавьте в список нового героя или измените порядок!

````python
# Герои сказки про репку
heroes = [
    "👵 Бабка",
    "👴 Дедка",
    "👩‍🌾 Внучка",
    "🐕 Жучка",
    "🐱 Кошка",
    "🐭 Мышка"
]

# Начинаем с бабки
chain = heroes[0]

# Постепенно добавляем героев
for hero in heroes[1:]:
    chain += " → " + hero

print("🌾 Кто тянул репку:")
print(chain)
print(f"\n📊 Всего героев: {len(heroes)}")
````

## 🐱 Переменные и типы данных

В Python переменная — это «коробка» для хранения данных.

**Основные типы данных:**
- `int` — целые числа (возраст, количество)
- `float` — дробные числа (вес, рост)
- `str` — текст (имена, описания)
- `bool` — логические значения (`True` или `False`)

**Обратите внимание:**
- Переменной просто присваивается значение через `=`
- Python автоматически определяет тип данных
- Используем `f-строки` для вставки переменных в текст: `f"Текст {переменная}"`

````python
def get_age_suffix(age):
    if age % 100 in [11, 12, 13, 14]:
        return "лет"
    if age % 10 == 1:
        return "год"
    elif age % 10 in [2, 3, 4]:
        return "года"
    else:
        return "лет"

# Кот Фунтик (экономист)
name = "Фунтик"        # str - текст 📝
age = 11                # int - целое 🔢
weight = 4.5           # float - дробное ⚖️
is_hungry = True       # bool - да/нет 🍽️

suffix = get_age_suffix(age)
print(f"🐱 {name}, возраст {age} {suffix}")
print(f"Вес: {weight} кг")
print(f"Голоден? {is_hungry}")
````

## 🧮 Арифметические операции

Python умеет выполнять все базовые математические операции.

**Базовые операции:**
- `+` — сложение
- `-` — вычитание
- `*` — умножение
- `/` — деление (всегда возвращает дробное число)

**Специальные операции:**
- `//` — **целочисленное деление** (только целая часть)
- `%` — **остаток от деления**
- `**` — **возведение в степень**

**Обратите внимание:**
- В примере используется форматирование `{per_day:.1f}` — выводит число с одним знаком после запятой

````python
# Фунтик считает свои расходы 💰

# Расходы за неделю
food = 100      # фунтов на еду
toys = 50       # фунтов на игрушки
vet = 200       # фунтов на ветеринара

# Общие расходы
total = food + toys + vet
print(f"💰 Всего расходов: {total} фунтов")

# Делим на количество дней (7)
per_day = total / 7
print(f"📊 В день: {per_day:.1f} фунтов")

# Если Фунтик съедает 50 г в день
grams_per_day = 50
days = 7
total_grams = grams_per_day * days
print(f"🍽️ За неделю: {total_grams} граммов корма")
````

## 🔀 Условные операторы (if, elif, else)

Условные операторы позволяют программе принимать решения.

**Структура:**
```python
if условие:
    # выполнится, если условие истинно
elif другое_условие:
    # выполнится, если первое ложно, а это истинно
else:
    # выполнится, если все условия ложны
```

**Операторы сравнения:**
- `==` — равно
- `!=` — не равно
- `<` — меньше
- `>` — больше
- `<=` — меньше или равно
- `>=` — больше или равно

**⚠️ Важно:** Не путайте `=` (присваивание) и `==` (сравнение)!

````python
# Возраст Фунтика 🐱
age = 2

if age < 1:
    print("🐱 Котёнок!")
elif age < 3:
    print("🐱 Молодой кот!")
elif age < 10:
    print("🐱 Взрослый кот!")
else:
    print("🐱 Мудрый кот!")

# Результат: 🐱 Молодой кот!
````

## 🔗 Логические операторы (and, or, not)

Логические операторы позволяют комбинировать несколько условий.

**Три основных оператора:**
- `and` — **И** (оба условия должны быть True)
- `or` — **ИЛИ** (хотя бы одно условие должно быть True)
- `not` — **НЕ** (переворачивает значение)

**Примеры:**
- `age > 1 and age < 5` — возраст от 2 до 4
- `weight < 3 or weight > 7` — вес либо меньше 3, либо больше 7
- `not is_hungry` — НЕ голоден

**Обратите внимание:** в примере оба кота удовлетворяют условиям (оба значения True), поэтому попадают в свои «факультеты».

````python
# 🎩 Магическая шляпа для котов

# Кот Байтик 🐱💻
print("🎩 Тестируем Байтика...")
loves_code = True
works_at_night = False

if loves_code and (not (not works_at_night)):
    print("🦁 ГРИФФИНДОР!")
    print("   Байтик — смелый программист! 💻")
    print("   🚀 Создаёт инновационные проекты!")
else:
    print("📚 Шляпа думает...")

# Кот Фунтик 🐱📊
print("\n🎩 Тестируем Фунтика...")
analyzes_markets = True
makes_forecasts = True

if analyzes_markets and makes_forecasts:
    print("🦅 КОГТЕВРАН-ЭКОНОМИКУС!")
    print("   Фунтик — мудрый экономист! 📊")
    print("   📈 Видит будущее в данных!")
else:
    print("📚 Шляпа думает...")
````

## 🔁 Цикл for

Цикл `for` используется для перебора элементов последовательности (списка, строки и т.д.).

**Структура:**
```python
for элемент in последовательность:
    # действия с элементом
```

**Обратите внимание:**
- `days` — это список дней недели
- Цикл `for` проходит по каждому элементу списка
- Внутри цикла можно использовать условия `if`
- Символ `\n` добавляет новую строку для лучшей читаемости

````python
# Фунтик учится в университете 🎓

days = ["Понедельник", "Вторник", "Среда",
        "Четвергиус", "Четверг", "Пятница", "Суббота", "Воскресенье"]

for day in days:
    if day in ["Суббота", "Воскресенье", "Четвергиус"]:
        print(f"🐱 {day}: Фунтик отдыхает! 😴")
        print(f"    Спит, ленится, зависает в интернете")
    else:
        print(f"🐱 {day}: Фунтик на паре! 📚")
        print(f"    Учит Data Science!")

print("\n📊 Неделя пройдена!")
````

## ⏱️ Цикл while

Цикл `while` выполняется, пока условие истинно.

**Структура:**
```python
while условие:
    # действия
    # не забудьте изменить условие, иначе будет бесконечный цикл!
```

**Обратите внимание:**
- Счётчик `attempts` увеличивается на каждой итерации
- Когда `attempts` станет равным 3, условие `attempts < 3` станет ложным
- Цикл остановится

**Бонус:** второй пример показывает обратный отсчёт с помощью `range(5, 0, -1)`

````python
# Байтик проверяет программу 💻

attempts = 0

while attempts < 3:
    print(f"🐱 Попытка {attempts + 1}")
    print(f"   Запуск программы...")
    attempts = attempts + 1

print("✅ Программа запустилась!")

print("\n" + "="*30)
````

````python
# Запуск ракеты! 🚀
print("🚀 ЗАПУСК!")
print("="*30)

for i in range(5, 0, -1):
    print(f"⏱️  {i}...")

print("="*30)
print("🚀 ВЗЛЁТ!!! 🌌")
````

## 📋 Списки (List)

Список — это упорядоченная коллекция значений.

**Важные особенности:**
- Индексы начинаются с **0** (первый элемент — `list[0]`)
- Можно обращаться с конца: `list[-1]` — последний элемент
- `append()` — добавить элемент в конец списка
- `len()` — получить количество элементов
- `enumerate()` — получить индекс и значение одновременно

**Обратите внимание:**
- В примере используется `enumerate(todo_list, 1)` — нумерация начинается с 1

````python
# Список дел Фунтика 📋

todo_list = [
    "📚 Написать эссе",
    "💻 Задание по Python",
    "🎬 Лекция на YouTube",
    "☕ Кофе с друзьями",
    "🎮 Видеоигры"
]

# Вывести весь список
print("🐱 Дела Фунтика:")
print(todo_list)

# Доступ по индексу
print("\nСамое важное:")
print(todo_list[0])

print("\nСамое приятное:")
print(todo_list[-1])

# Добавить новое дело
todo_list.append("🏋️ Спортзал")

# Количество дел
count = len(todo_list)
print(f"\n🐱 Всего дел: {count}")

# Цикл по списку (первые 3)
print("\nПервые 3 дела:")
for i in range(3):
    print(f"  {i+1}. {todo_list[i]}")
````

````python
# Байтик проверяет завтрак 🥄
mice = 3  # мышки на завтрак
sausages = 1  # сосиски

if mice >= 3:
    print("✅ Отличный завтрак!")
elif mice >= 1 and sausages > 0:
    print("⚠️ Нормальный завтрак")
else:
    print("❌ Где мой завтрак?!")
    print("Я требую завтрак!")
````

## 🐱 Функции — многократное использование кода

Функция — это блок кода с именем, который можно вызывать многократно.

**Зачем нужны функции?**
- Не повторять один и тот же код
- Сделать программу понятнее
- Легко вносить изменения

**Структура:**
```python
def имя_функции():
    # код функции
    print("Что-то делаем")

# Вызов функции
имя_функции()
```

**Обратите внимание:** вместо 6 строк `print` мы пишем одну строку `introduce()`!

````python
# Создаём функцию представления
def introduce():
    print("="*50)
    print("🐱 Привет, я кот Байтик!")
    print("📍 Студент 2 курса")
    print("🏫 Изучаю Data Science")
    print("💻 Язык: Python")
    print("="*50)

# Вызываем функцию один раз
introduce()
````

## 🔧 Функции с параметрами

Параметр — это значение, которое функция принимает для работы.

**Важные термины:**
- **Параметр** — переменная в определении функции: `def add_food(amount):`
- **Аргумент** — конкретное значение при вызове: `add_food(50)`
- **global** — ключевое слово для использования глобальной переменной внутри функции

**Обратите внимание:**
- `total_food` объявлена вне функции (глобальная переменная)
- Внутри функции используем `global total_food` для её изменения
- Каждый вызов функции передаёт разное значение: 50, 70, 60

````python
# Фунтик считает, сколько он съел за день 🍽️

total_food = 0  # Переменная для накопления еды

def add_food(amount):
    global total_food
    total_food = total_food + amount
    print(f"🐱 Фунтик съел {amount} г. Всего: {total_food} г")

# Завтрак
add_food(50)

# Обед
add_food(70)

# Ужин
add_food(60)

# Финальный результат
print(f"\n📊 Итого за день: {total_food} граммов")
````

## 💰 Функции с несколькими параметрами

Функция может принимать не один, а несколько параметров.

**В этом примере:**
- Два параметра: `amount` (сумма) и `name` (описание операции)
- Условие внутри функции: проверяем, положительная или отрицательная сумма
- Используем разные эмодзи для доходов (✅) и расходов (❌)

**Обратите внимание:** функция сама решает, как форматировать вывод в зависимости от знака числа!

````python
# Баланс Байтика в биткойнах ₿

balance = 0

def transaction(amount, name):
    global balance
    balance = balance + amount

    if amount > 0:
        print(f"✅ {name}: +{amount}₿")
    else:
        print(f"❌ {name}: {amount}₿")

    print(f"   Баланс: {balance}₿\n")

# Транзакции
transaction(100, "Зарплата")
transaction(-30, "Корм")
transaction(-20, "Игрушки")
transaction(50, "Премия")

print(f"💰 Итого: {balance}₿")
````
